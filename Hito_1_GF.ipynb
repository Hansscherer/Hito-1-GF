# Sector

La empresa elegida fue Advanced Micro Devices, o comúnmente llamada AMD. Es una empresa estadounidense de semiconductores dedicada específicamente a la fabricación de microprocesadores X86 de computación y otros productos tecnológicos similares.

AMD se fundó en 1969 por un grupo de ejecutivos de Fairchild Semiconductores, la que fue la primera empresa en comerciar circuitos integrados. En sus inicios, empezó produciendo memorias RAM y procesadores para computadoras, siendo Intel su principal competidor. En 2006 adquirió ATI, empresa dedicada a producir tarjetas gráficas, por lo que después de esta adquisición AMD pasó a ser competencia de NVIDIA, la empresa más importante de tarjetas gráficas.

En el último tiempo, AMD se ha enfocado en la arquitectura Zen, la cual ha demostrado un rendimiento sobresaliente respecto a su arquitectura antecesora. Con el desarrollo de esta nueva arquitectura, AMD colaboró para confeccionar dispositivos como Steam Deck, PlayStation 5, Xbox X y S.

La historia dice que Intel ha sido la empresa líder de este mercado de microprocesadores X86 para computadoras. De hecho, en el paper de Semmler (2010), llamado "*Competition in the microprocessor market: Intel, AMD and Beyond*" Intel puede definirse como un oligopolio, siendo la empresa líder y AMD la seguidora. Entre ambas compañías poseen prácticamente el 100% de la cuota de mercado de los procesadores X86 comerciales de la actualidad, dividiendo la cuota en un 23,9% para AMD y el 76,1% restante para Intel.

Las características típicas de este oligopolio, según Semmler (2010), son:

  1. Las empresas venden productos estandarizados.
  2. Las empresas establecen precios.
  3. Existen grandes barreras de entrada.
  4. Hay pocos competidores y hay interacción entre ellas.
  5. La publicidad, dominación de la marca y la tecnología diferencian el producto.

Según Rodríguez y Vargas, en el paper "*Ventaja competitiva de Intel en el segmento de los microprocesadores*" (2020), las barreras de entrada de este mercado son bastante altas. Por ejemplo, las alianzas con fabricantes de PCs, el alto costo de creación y mantenimiento de las fábricas, y el rápido desarrollo de la tecnología. Es por lo anterior que solo existen dos empresas importantes en este mercado.

En el mercado de tarjetas gráficas, NVIDIA es el amplio dominador con el 88% de participación, y AMD con el 12% restante. Intel intentó entrar en ese mercado en 2023, obteniendo un 4% de participación, pero fracasó y en la actualidad no tiene participación de mercado. Mientras que AMD ha mantenido su 12%, NVIDIA absorbió la parte de Intel, siendo el dominador absoluto del apartado gráfico de microprocesadores.

Para valorar esta opción de compra se utilizará un **Árbol Binomial**, ya que permite valorar la opción y adicionalmente comparar esta inversión con otros activos en las mismas condiciones. En este caso, además de valorar la opción de compra de la acción de AMD, se evaluará un Bono del Tesoro de Estados Unidos ($B$) a 1 año con una tasa del 3,95% ($r$). Este bono fue elegido a este plazo, porque estimo que a mediano plazo la acción de AMD puede tener un crecimiento importante, pero a 5 o 10 años me parece mucho tiempo al paso que avanza la tecnología e Intel podría retomar su absoluto liderazgo histórico.



El precio de la acción de AMD ($S_0$) el 19 de septiembre de 2024 es de 156,74 dólares, el precio de ejercicio ($K$) es 55 dólares, la fecha de expiración es el 19 de septiembre de 2025, el multiplicador del precio alto ($u$) será 1,5, el de precio bajo ($d$) 0,6, y el valor del bono es de 100 dólares.

Para poder empezar con los cálculos, se necesita la probabilidad de que ocurra $u$ y $d$, por lo que a continuación se harán los cálculos correspondientes:

```python
from sympy import *
init_printing()

# Parámetros Bono
r, B = symbols('r B')
# Parámetros Acción
u, d, S0 = symbols('u d S0')

W = Matrix([[-B, -S0], [r*B, u*S0],[r*B, d*S0]])
W
```

El Teorema Fundamental de Valoración de Activos (Harrison and Kreps (1979)) establece que la estructura financiera es libre de arbitraje si y sólo si existe un  $\pi\in\mathbb{R}^S_{++}$ tal que

(Condición)  $[1,\pi']W=0$.

Para utilizar este resultado, definamos el vector  $[1,\pi']$.

```python
pi_u, pi_d = symbols('pi_u pi_d')
pi = Matrix([[1],[pi_u],[pi_d]])
pi
```

Reemplazando en la (Condición) obtenemos el siguiente sistema de ecuaciones:

```python
NA=pi.transpose()*W
NA
```

Luego con solver se pueden encontrar las probabilidades $\pi_1$ y $\pi_2$

```python
piNA=solve([NA[0],NA[1]],[pi_u,pi_d],dict=True)
piNA
```

Ahora se reemplazan los parámetros en la ecuación obtenida:

```python
from sympy import Matrix

def pi_u(d,r,u):
    return (-d+r)/(-d*r+r*u)

def pi_d(d,r,u):
    return (-r+u)/(-d*r+r*u)

# Parámetros
r = 1.0395  # Tasa libre de riesgo a 1 año
u = 1.5     # multiplicador up
d = 0.6     # multiplicador down

# Cálculo de probabilidades
pi_u_value = pi_u(d,r,u)
pi_d_value = pi_d(d,r,u)

# Crear la matriz
piNA = Matrix([[pi_u_value],[pi_d_value]])
piNA
```

Ahora que ya se tienen las probabilidades de que ocurra $\pi_1$ y $\pi_2$, se puede seguir con el cálculo y reemplazar los parámetros en el modelo binomial.

```python
from sympy import *
# Parámetros Opción
q3 = 99.83
# Parametros
S0 = 156.74 #Valor accion AMD 19-09-2024
K = 55 #Precio de ejercicio opcion
r = 1.0395 #Tasa libre de riesgo a 1 año
B = 100
u = 1.5  #multiplicador up
d = 0.6  #multiplicador down

W = Matrix([[-B, -S0, -q3], [r*B, u*S0, max(0,u*S0-K)],[r*B, d*S0, max(0,d*S0-K)]])
W
```
